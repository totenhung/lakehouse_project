```python
df = spark.read.option("header",True).csv(f_name)
```

```python
# Drop Duplicates
df1 = df.dropDuplicates()
# Drop Null Values
df1 = df1.na.drop()
```

```python
from pyspark.sql.functions import split, col
from pyspark.sql.types import ArrayType, StringType

# Định nghĩa hàm UDF (User-Defined Function) để chuyển đổi chuỗi thành mảng
# Tách thể loại của nghệ sĩ ra thành nhiều dòng
split_udf = split(col("genres"), ", ")
# Đổi cột với kiểu dữ liệu là ArrayType
df2 = df1.withColumn("genres", split_udf.cast(ArrayType(StringType())))
# Explode multiple values in artist ids

from pyspark.sql.functions import explode
df1 = df2.withColumn("genres",explode(df2.genres))
```

```python
df2 = df1.withColumn("popularity",df1.popularity.cast('float'))
df1 = df2
df2 = df1.withColumn("followers",df1.followers.cast('int'))
```

```python
%%sql
CREATE DATABASE IF NOT EXISTS silver
```

```python
%%sql
DROP TABLE IF EXISTS silver.spotify_artists
```

```python
%%sql
CREATE TABLE IF NOT EXISTS silver.spotify_artists(
    artist_id	              string,
    artist_name            string,
    popularity  float,
    genres string,
    followers int,
    external_urls string
    )
USING iceberg
PARTITIONED BY (genres)
```

```python
df2.writeTo("silver.spotify_artists").append()
```
